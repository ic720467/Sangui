# PRYECTO DE FRACTALES

## Fractal 1



$z=z^6+C$

```python
import numpy as np
import scipy as sp 
import matplotlib.pyplot as plt
%matplotlib inline
#definir la función 
def fun1(z,c):
    return z**6+c

x,y=np.ogrid[-2:1:1000j,-1.5:1.5:1000j]
c= x+1j*y
z=0
iterar=300

for i in range(iterar):
    z=fun1(z,c)
    
threshold=2
mask=np.abs(z)<threshold
plt.imshow(mask.T,extent=[-2,1,-1.5,1.5])
plt.gray()
plt.show()
```

## Fractal 2

### $$z=z ^ 2 + c$$

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

def fun1 (z,c):
    return z ** 2 + c

x,y=np.ogrid[-2:1:1000j,-1.5:1.5:1000j]
c = x + 1j*y
z=0
iterar = 300

for i in range(iterar):
    z = fun1(z,c)
    
threshold = 2
mask=np.abs(z) < threshold
plt.imshow(mask.T,extent=[-2,1,-1.5,1.5])
plt.gray()
plt.show()
```

## Fractal 3

### $$z ^ 2 + \frac{1}{c}$$

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline

def fun1 (z,c):
    return z ** 2 + 1/c

x,y=np.ogrid[-2:4:1000j,-2:2:1000j]
c = x + 1j*y
z=0
iterar = 300

for i in range(iterar):
    z = fun1(z,c)
    
threshold = 2
mask=np.abs(z) < threshold
plt.imshow(mask.T,extent=[-2,4,-2,2])
plt.gray()
plt.show()
```

# PROYECTO DE ANÁLISIS DE DATOS DE LA NASA 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

pd.set_option('display.max_columns', 7)

pd.set_option('display.max_rows', 10)

pd.set_option('precision', 3)
```

```python
temp = pd.read_csv('FIRAS_TEMPERATURE_MAP_LOWF_t1.csv')
```

```python
temp
```

```python
t = pd.DataFrame(temp)
min = np.argmin(t["TEMP"])
max = np.argmax(t["TEMP"])
```

```python
min = np.argmin(t["TEMP"])
max = np.argmax(t["TEMP"])
t[["TEMP"]].plot()
plt.plot(min,t["TEMP"][min],'mo')
plt.plot(max,t["TEMP"][max],'ro')
plt.title('Gráfica Latitud vs Longitud, Temperatura, Tamaño', fontsize = 16, weight = 'bold',family = 'serif')
plt.xlabel('$Longitud$', fontsize = 18)
plt.ylabel('$Latitud$', fontsize = 18)
plt.show()
```

```python
#DEFINIMOS UNA FUNCIÓN PARA MARACR LA ESCALA DE LA TEMPERATURA EN BASE A COLORES
def barra_colores():
    from matplotlib import pyplot
    import matplotlib as mpl
#COLOCAMOS EL RANGO DE TEMPERATURA MÍNIMA Y MÁXIMA PARA LA ESCALA DE COLORES
    num_color_min = int(t['TEMP'][min]*100000000 - 270000000)
    num_color_max = int(t['TEMP'][max]*100000000 - 270000000)

    barra_color = np.arange(num_color_min, num_color_max, 100000)
    hex_colors = []
    bounds = []
#ASIGNAMOS LOS COLORES Y LA POSOCIÓN DE LA ESCALA EN FORMA HORIZONTAL 
    for i in range(17):
        hex_colors.append('#' + "%06x" % int(barra_color[i]))
        barra_color[i] = int((barra_color[i] + 270000000)/100000)
        bounds.append(int(barra_color[i]))

    fig = pyplot.figure(figsize=(8, 3))
    ax1 = fig.add_axes([0.0, 0.0, 1.5, 0.2])

    cmap = mpl.colors.ListedColormap(hex_colors)

    norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
    cb2 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                    norm=norm,
                                    ticks=bounds,  
                                    spacing='proportional',
                                    orientation='horizontal')
    cb2.set_label('Temperatura de CMBR (Radiación de fondos cósmicos de microondas ) [mK]',size=14)
    return pyplot.show()
```

```python
#TOMAMO LA LATITUD Y LONGUITUD  DE NUESTRA SERIE DE DATOS PARA GRAFICAR LA ESFERA CELESTE
#ADEMAS DE AGREEGAR EL AUMENTO DE TAMAÑO EN LOS PUNTOS EN BASE A SU PESO
fig,axes = plt.subplots(1, 1, figsize=(20,10))
for i in range(6067):
    num_color = int(t['TEMP'][i]*100000000 - 270000000)
    hex="%06x" % num_color
    hex = '#' + hex
    tam = t['WEIGHT'][i]
    if tam < 1:
        tam += 1
    plt.plot(t['GAL_LON'][i], t['GAL_LAT'][i], 'o', ms = tam*2, color=hex)
plt.plot(t['GAL_LON'][min], t['GAL_LAT'][min], 's', ms = 10, color ='m',label='Minimo')
plt.plot(t['GAL_LON'][max], t['GAL_LAT'][max], 's', ms = 12, color ='r',label='Maximo')
plt.xlim(0,360)
plt.title('Gráfica Latitud vs Longitud, Temperatura, Tamaño', fontsize = 16, weight = 'bold',family = 'serif')
plt.xlabel('$Longitud$', fontsize = 18)
plt.ylabel('$Latitud$', fontsize = 18)
plt.legend()
plt.show()
barra_colores()
```

```python
fig,axes = plt.subplots(1, 1, figsize=(20,10),subplot_kw={'projection': 'aitoff'}) 
for i in range(6067): 
    num_color = int(t['TEMP'][i]*100000000 - 270000000) 
    hex="%06x" % num_color 
    hex = '#' + hex 
    tam = t['WEIGHT'][i] 
    lat = t['GAL_LAT'][i] 
    if tam < 1: 
        tam += 1 
    plt.plot(np.radians(t['GAL_LON'][i] - 180), np.radians(lat),'bo', ms = tam*2, color=hex) 
plt.plot(np.radians(-t['GAL_LON'][min]), np.radians(t['GAL_LAT'][min]), 'o', ms = 10, color ='y') 
plt.plot(np.radians(t['GAL_LON'][max] - 180), np.radians(t['GAL_LAT'][max]), 'o', ms = 10, color ='r') 
plt.title('Gráfica Latitud vs Longitud, Temperatura, Tamaño', fontsize = 16, weight = 'bold',family = 'serif') 
plt.show() 
barra_colores()
```
